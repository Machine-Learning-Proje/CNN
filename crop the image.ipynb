```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

img = cv2.imread(r"C:\Users\USER\Desktop\our project\axial\Alzheimer_s Dataset\train\NonDemented\nonDem0.jpg",0)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)

```

```python

th = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

ret,th = cv2.threshold(th, 110, 255, cv2.THRESH_BINARY)

kernel = np.ones_like((5,5),np.uint8)

erosion = cv2.erode(th,kernel,iterations = 1)

# dilation = cv2.dilate(erosion,kernel,iterations = 1)

plt.imshow(erosion, 'gray')





```

```python
cont,_= cv2.findContours(erosion, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

```

```python
# contourImg=cv2.drawContours(img,cont,-1,(255,0,0),1)
# plt.imshow(contourImg)
```

```python
# find the biggest countour (c) by the area

c = max(cont, key = cv2.contourArea)

x,y,w,h = cv2.boundingRect(c)



# draw rectangle on (c) in green

cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),0) 

plt.imshow(img)


```

```python
cropped_imge= img[y:y+h, x:x+w]
plt.imshow(cropped_imge,'gray')
```

```python
IMG_SIZE_y = 100
IMG_SIZE_x = 85
resiz = cv2.resize(cropped_imge,(IMG_SIZE_x,IMG_SIZE_y))
plt.imshow(resiz,'gray')
print (resiz.shape)
```
